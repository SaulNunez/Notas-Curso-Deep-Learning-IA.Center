# Procesamiento de imagenes

## OpenCV

```python
import cv2

img_path = "/content/RGB-Profile-Color-Wheel.png"

image = cv2.imread(img_path)
```

```python
from google.colab.patches import cv2_imshow

cv2_imshow(image)
```

```python
B, G, R = cv2.split(image)

cv2_imshow(B)
cv2_imshow(G)
cv2_imshow(R)
```

## Eye

```python
eye_path = "/content/eye.png"

eye = cv2.imread(eye_path)
```

```python
cv2_imshow(eye)

eye_b, eye_g, eye_r = cv2.split(eye)

cv2_imshow(eye_r)
cv2_imshow(eye_g)
cv2_imshow(eye_b)
```
