# Ejercicio 2

Cargar imagenes

```python
import cv2

fruits = cv2.imread('/content/frutas.png.png')
```

```python
from google.colab.patches import cv2_imshow

cv2_imshow(fruits)
```

## Imagen 1

Generar la imagen en escalade grises.

```python
fruits_gray = cv2.cvtColor(fruits, cv2.COLOR_BGR2GRAY)

cv2_imshow(fruits_gray)
```

## Image 2

Generar una función para separarla imagen en los 3 canales BGR y obtener una imagen para cada canal de surespectivo color activo.

```python
fruits_b = fruits.copy()

# Blue
fruits_b[:,:, 1] = 0
fruits_b[:,:, 2] = 0

cv2_imshow(fruits_b)

fruits_g = fruits.copy()

# Green
fruits_g[:,:, 0] = 0
fruits_g[:,:, 2] = 0

cv2_imshow(fruits_g)

fruits_r = fruits.copy()

# Red
fruits_r[:,:, 0] = 0
fruits_r[:,:, 1] = 0

cv2_imshow(fruits_r)
```

## Image 3

Sumar los 3 canales de la imagen para obtener la imagen original.

```python
cv2_imshow(fruits_r + fruits_b + fruits_g)
```
